```python
df1 = df.copy()
df1['name'] = df1['name'].str.replace('ё', 'е').str.strip().str.replace('\xa0', ' ').str.lower()
```

```python
idx_ignore = []
```

```python
for locality in okato_type:
    df_filtered = df1[df1['name'].str.contains(f'{locality} ', case=False)].copy()
    if len(df_filtered) != 0:
        idx = list(set(df_filtered.index) - set(idx_ignore))
        df1.loc[idx, 'type'] = locality
        df1.loc[idx, 'only_name'] = df_filtered['name'].str.split(f'{locality} ').str[-1]
        
        idx_ignore.extend(df_filtered.index.tolist())
```
